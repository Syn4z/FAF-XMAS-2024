```python
import numpy as np

def haversine_distance_vectorized(coords1, coords2):
    """
    Calculate the straight-line distances (in meters) between multiple geographic points using the Haversine formula.
    
    Args:
        coords1: Array of shape (n, 2) where each row is (lat, lon) in decimal degrees.
        coords2: Array of shape (n, 2) where each row is (lat, lon) in decimal degrees.

    Returns:
        Distances in meters as a 1D NumPy array.
    """
    R = 6378137  # Radius of Earth in meters

    # Convert latitude and longitude to radians
    lat1, lon1 = np.radians(coords1[:, 0]), np.radians(coords1[:, 1])
    lat2, lon2 = np.radians(coords2[:, 0]), np.radians(coords2[:, 1])

    # Compute differences
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    # Haversine formula
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    # Distance in meters
    distances = R * c
    return distances


distances = np.zeros((len(coordinates), len(coordinates)))
# calculate the distance between all the coordinates
for i in range(len(coordinates)):
    distances[i] = haversine_distance_vectorized(np.array([coordinates[i]]), np.array(coordinates))

print(distances)
```

```python
import folium
import numpy as np

class RouteVisualizer:
    def __init__(self, df=None, ordered_list=None):
        self.df = df
        self.ordered_list = ordered_list

    def plot_route(self):
        """
        Plot the route on a map using Folium.
        """
        # Get the coordinates in the correct order (latitude, longitude)
        coords = self.df[['X Coordinate', 'Y Coordinate']].values  # Switch to (latitude, longitude)
        coords = coords[self.ordered_list]

        # Get the center of the map (latitude, longitude)
        center = np.mean(coords, axis=0)

        # Create the map
        m = folium.Map(location=center, zoom_start=10)  # Adjust zoom_start for Spain

        # Add markers for each location
        for i, coord in enumerate(coords):
            popup_content=f"Point: {i} \nName: {self.df.iloc[i]['Name']} \nConclusion: Good Child",
            folium.Marker(
                location=coord,
                popup=folium.Popup(f"Point: {i} <br> Name: {self.df.iloc[i]['Name']} <br> Conclusion: Good Child", max_width=150, font_size=20),   
                icon=folium.Icon(color="blue" if i not in (0, len(coords) - 1) else "green" if i == 0 else "red", icon="gift"),
            ).add_to(m)

        # Add lines connecting the points
        folium.PolyLine(coords, color="red", weight=2.5, opacity=1).add_to(m)

        return m
```

```python
route_visualizer = RouteVisualizer(df_germany, ordered_route)
m = route_visualizer.plot_route()
m.save('route_germany.html')
```
